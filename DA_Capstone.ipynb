```python
# Task 1: Percentage of road accidents during all the years

# Sum the total number of accidents for each year (2013 to 2016)
total_accidents_per_year = df.iloc[:, 2:6].sum()

# Calculated the overall total accidents across all years
total_accidents_all_years = total_accidents_per_year.sum()

# Calculated the percentage of accidents for each year
accident_percentage_per_year = (total_accidents_per_year / total_accidents_all_years) * 100

# Display the percentage of accidents per year
accident_percentage_per_year
```

```python
# Task 2: Mean Accidents per 1L population

# Columns for accidents per 1 lakh populatin per year 

# Extracted accidents per 10,000 vehicles for each year
accidents_per_1L_population_2013 = df['Total Number of Accidents Per Lakh Population - 2013']
accidents_per_1L_population_2014 = df['Total Number of Accidents Per Lakh Population - 2014']
accidents_per_1L_population_2015 = df['Total Number of Accidents Per Lakh Population - 2015']
accidents_per_1L_population_2016 = df['Total Number of Accidents Per Lakh Population - 2016']

# Calculated the mean accidents per 1 lakh populatin for each year
mean_accidents_2013 = accidents_per_1L_population_2013.mean()
mean_accidents_2014 = accidents_per_1L_population_2014.mean()
mean_accidents_2015 = accidents_per_1L_population_2015.mean()
mean_accidents_2016 = accidents_per_1L_population_2016.mean()

# Display the results
print(f"Mean accidents per 1L population in 2013: {mean_accidents_2013}")
print(f"Mean accidents per 1L population in 2014: {mean_accidents_2014}")
print(f"Mean accidents per 1L population in 2015: {mean_accidents_2015}")
print(f"Mean accidents per 1L population in 2016: {mean_accidents_2016}")
```

```python
# Task 3: Highest and least number of accident states

# Filter out the row where the 'States/UTs' column has "Total"
df = df[df['States/UTs'] != 'Total']

# Summing the total number of accidents for each state from 2013 to 2016
df['Total_Accidents'] = df.iloc[:, 2:6].sum(axis=1)

# Find the state with the highest number of accidents
highest_accident_state = df.loc[df['Total_Accidents'].idxmax(), 'States/UTs']
highest_accident_value = df['Total_Accidents'].max()

# Find the state with the least number of accidents
lowest_accident_state = df.loc[df['Total_Accidents'].idxmin(), 'States/UTs']
lowest_accident_value = df['Total_Accidents'].min()

# Display the results
print(f"The state with the highest number of accidents is {highest_accident_state} with {highest_accident_value} accidents.")
print(f"The state with the least number of accidents is {lowest_accident_state} with {lowest_accident_value} accidents.")
```
